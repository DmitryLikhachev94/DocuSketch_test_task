# Inference and further exploration

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('darkgrid')
```

```python
df = pd.read_json('deviation.json', encoding='cp1251')
df.sample(5)
```

The mean column is the average value of floor_mean and ceiling_mean, similarly for the max and min columns.
Floor_mean is not just an average value between floor_max and floor_min. Similarly for ceiling columns.
Therefore, with the help of a little research on the topic, we conclude that there were several measurements for the angles of the ceiling and floor corners.

```python
df.name.nunique()
```

There are to many different room names, so this column is not useful. Also, the gt_corners is the same as rb_corners, so one of them is enough.

```python
df.drop(['name', 'rb_corners'], axis=1, inplace=True)
```

Lets check the amount of corners in our rooms.

```python
df.gt_corners.unique()
```

```python
df.gt_corners.value_counts()
```

```python
df_grouped = df.groupby('gt_corners').mean()
df_grouped
```

```python
sns.heatmap(df_grouped, annot=True)
```

```python
fig, axs = plt.subplots(3,3, figsize=(18,10))
for indx, col in enumerate(df.columns[1:]):
    row_idx = indx // 3
    col_idx = indx % 3
    ax = axs[row_idx, col_idx]
    sns.boxplot(x=df['gt_corners'], y=col, data=df, ax=ax)
```

As we can see, the more corners in the room, the more average deviation values in degrees.
